```python
import pandas as pd 
import numpy as np  
import seaborn as sns  
import matplotlib.pyplot as plt 
import plotly.express as px 
```

```python
rent_df = pd.read_csv('House_Rent_Dataset.csv')
```

```python
rent_df.head()
```

### BARCHARTS

```python
# 1. Create a barchart of average rent vs city

sns.barplot(data= rent_df, x = 'City' ,y = 'Rent')
plt.title('Average Rent Over City')
```

```python
# 2. Create a barchart of sum of rent  vs bathroom
# calculate the sum of rent for each bathroom
sum_df = rent_df.groupby('Bathroom')['Rent'].sum().reset_index()
# plot the barchart
sns.barplot(data = sum_df, x ='Bathroom', y ='Rent')
plt.title('Sum Of Rent Over Number of Bathroom')
```

```python
# 3. create a bar chart of average rent vs area type
sns.barplot(data = rent_df, x = 'Area Type', y = 'Rent')
plt.title('Average Rent Over Area Type')
```

```python
# 4.create a bar chart of average rent vs furnishing status

sns.barplot(data = rent_df, x ='Furnishing Status', y= 'Rent')
plt.title('Average Rent Over Furnishing Status')
```

```python
# 5. Create a bar chart of size vs city

sns.barplot(data = rent_df, x ='City', y ='Size')
plt.title('Size Over City')
```

### LINE CHART

```python
# 6. Create a line chart of average rent vs posted on
# convert posted on to datetime
rent_df['Posted On'] = pd.to_datetime(rent_df['Posted On'])
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.lineplot(data = rent_df, x ='Posted On', y ='Rent', marker = 'o')
# title
plt.title('Average Rent Over Time')
```

### Scattered Chart

```python
# 7. Create a scattered chart of rent vs size
sns.scatterplot(data = rent_df, x ='Size', y ='Rent', alpha=0.7)
plt.title('Rent Over Size')
plt.figure(figsize=(8,6))
```

```python
# 8. Create a scattered chart of sum of rent vs no of bathroom
sns.scatterplot(data = rent_df, x ='Bathroom', y ='Rent', alpha=0.7)
plt.title('Rent Over No of Bathroom ')
plt.figure(figsize=(8,6))
```

```python
# 9. Create a scattered chart of rent vs city
sns.scatterplot(data = rent_df, x ='City', y ='Rent', alpha=0.7)
plt.title('Rent Over City ')
plt.figure(figsize=(8,6))
```

```python
# 10. Create a scattered chart of sum of rent vs area locality
rent_df['Posted On'] = pd.to_datetime(rent_df['Posted On'])
sns.scatterplot(data = rent_df, x ='Posted On', y ='Size', alpha=0.7)
plt.title('Size Over Posted On')
plt.figure(figsize=(10,6))
```

```python
# 11. Create a scattered chart of sum of rent vs area locality
sns.scatterplot(data = rent_df, x ='Size', y ='Rent', alpha=0.7)
plt.title('Rent Over Size')
plt.figure(figsize=(10,6))
```

### Violin Plot

```python
# 12.
sns.violinplot(data = rent_df, x = 'Area Type', y ='Rent', inner ='quartile')
plt.title('Distribution of Rent over Area Type')
```

### Heat Map

```python
# 13
# pivot the dataframe to have furnishing status as column,city as index,and rent as values
pivot_df = rent_df.pivot_table(index='City',columns='Furnishing Status', values='Rent',aggfunc='mean')
plt.figure(figsize=(10,6))
sns.heatmap(data = pivot_df, annot=True ,cmap='coolwarm', fmt='.1f', linewidths =.5)
plt.title('Rent Distribution by Furnishing Status and City')
```

### Histogram

```python
# 14. What is the distribution of the number of bathroom in houses
plt.figure(figsize=(10,6))
sns.histplot(data = rent_df['Bathroom'], bins= 10, kde= False)
plt.title('Distribution of Bathrooms')
```

### Bar Chart

```python
# 15. Distribution of houses by finishing status
plt.figure(figsize=(10,8))
furnishing_counts = rent_df['Furnishing Status'].value_counts()
plt.pie(furnishing_counts, labels=furnishing_counts.index,autopct='%1.1f%%',startangle=140)
plt.title('Distribution of Houses by Furnishing Status')
```

### Box Plot

```python
# 16. How does the rent distribution vary with the number of bathroom
sns.boxplot(data=rent_df, x= 'Bathroom', y='Rent')
plt.title('Distribtion of Rent with Numbers of Bathroom')
```

```python
# 17. What is the variation in rent price across the different cities
sns.boxplot(data=rent_df, x= 'City', y='Rent')
plt.title('Distribtion of Rent across the different Cities')
```

```python
# 18. How does area type affects rent
sns.boxplot(data=rent_df, x= 'Area Type', y='Rent')
plt.title('Distribtion of Rent with Area Type')
```

```python
# 19. What is the distribution of rent prices of house with different furnishing status
sns.boxplot(data=rent_df, x= 'Furnishing Status', y='Rent')
plt.title('Distribtion of Rent with Furnishing Status')
```

```python
# 20. How does the rent vary in size
sns.boxplot(data=rent_df, x= 'BHK', y='Rent')
plt.title('Distribtion of Rent with BHK')
```
